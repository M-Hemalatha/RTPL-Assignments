```python
no_of_fruits= [44,56,23,10,45]
fruits = ['apple','orange','banana','watermelon','strawberry']
plt.pie(no_of_fruits,labels = fruits)
plt.show()
```

```python
# By default the plotting of the first wedge starts from the x-axis and move counterclockwise:
# but you can change the start angle by specifying a startangle parameter.
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
#if you want one of the wedges to stand out use 'explode'
fruit_exp=np.array([0.2,0,0,0,0])
# must be an array with one value for each wedge.
# Each value represents how far from the center each wedge is displayed:
plt.pie(no_of_fruits,labels=fruits,explode=fruit_exp)
plt.show()
```

```python
# to add shadow enable  shadot to True
plt.pie(no_of_fruits,labels=fruits,explode=fruit_exp,shadow=True)
plt.show()
```

```python
#we can set our custom color for each wedge
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y,labels=mylabels,colors=mycolors,shadow=True)
plt.show()
```

```python
#we can also show the legend
plt.pie(no_of_fruits,labels=fruits)
plt.legend()
plt.show()
```

```python
# to add header to legend add title to legend
plt.pie(no_of_fruits,labels=fruits)
plt.legend(title="Fruits")
plt.show()
```

```python
#Advance Visualization
import numpy as np
import matplotlib.pyplot as plt

#defining dataset
x=np.arange(0,10,1)
y=2*x+5
z=3*x+8

#changing the size figure
fig = plt.figure(figsize=(10,5))

#Plotting datapoints
plt.plot(x,y,linewidth = 2.0,linestyle=':', color='r',alpha=0.8,marker='o')
plt.plot(x,z,linewidth=2.0,linestyle='--',color='g',alpha=0.5,marker='X')

plt.title("Line Plot demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','line2'],loc='best')
plt.grid()
plt.show()
```

```python
#Creating the dataset
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x =[1,2,3,4,3,2,1,2]

#Creating the scatter Plot
plt.scatter(a,b, c='g', s=200,edgecolors='y',marker='o',alpha =0.5)
plt.scatter(a,x,c='r', s=400,edgecolors='b',marker='^',alpha =1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
#plt.savefig('F://Django//udemycoupenscraper//udemycoupenscrapper//scraperapi//Scatter Plot.png')  #we can save any plot by using plt.savefig()
plt.show()
```

```python
#Other advanced charts
#box plot
#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.boxplot(data, showmeans = True)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```

```python
#violin plot
#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

```python
#doughnet plot
#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names, radius = 1.5,colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_centre,radius = 1.0,colors = 'w')
plt.show()
```

```python
#area chart

#define the dataset
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors = 'Black', alpha = 0.5)
#Merge area chart with the line chart
plt.plot(x,y,color = 'g')
plt.grid(True)
plt.show()
```
